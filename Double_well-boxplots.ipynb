```python
import matplotlib.pyplot as plt
import numpy as np

# OU process

LSTM_iter = [3050,8100,7850,8450,9050,8150,6450,8950,10400,3800]
LSTM_time = [340,910,880,943,1010,910,720,998,1161,424]

GRU_iter = [6000,4250,4450,5850,1950,6400,5300,5400,3700,4350]
GRU_time = [660,487,508,668,223,732,606,617,423,497]

Transformer_iter = [7250,5350,8750,5700,4450,5100,4250,6800,3900,4900]
Transforme_time = [1765,1296,2122,1381,1080,1236,1040,1650,957,1189]

niter=[LSTM_iter,GRU_iter,Transformer_iter]
time=[LSTM_time,GRU_time,Transforme_time]
labels =['LSTM','GRU','Transformer']

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

bplot1 = ax1.boxplot(niter,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('(a) Number of Iterations')

bplot2 = ax2.boxplot(time,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('(b) Total Computational Time')

# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Three architectures')
    ax.set_ylabel('Observed values')

ax1.set_ylabel('epoch')
ax2.set_ylabel('seconds')
plt.show()
```
