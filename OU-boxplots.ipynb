```python
import matplotlib.pyplot as plt
import numpy as np

# OU process

LSTM_iter = [650,500,750,550,650,950,600,750,750,750]
LSTM_time = [12.7,9.82,14.6,10.7,12.7,18.4,11.7,14.6,14.6,14.6]

GRU_iter = [400,500,500,400,700,250,500,350,600,650]
GRU_time = [8.88,11,11,8.84,15.4,5.52,11,7.74,13.2,14.3]

# Transformer_iter = [150,100,150,150,100,250,150,100,150,100]
# Transforme_time = [22.4,15,22.4,22.4,15,37,22.3,15,22.3,15]

# niter=[LSTM_iter,GRU_iter,Transformer_iter]
# time=[LSTM_time,GRU_time,Transforme_time]
# labels =['LSTM','GRU','Transformer']
niter=[LSTM_iter,GRU_iter]
time=[LSTM_time,GRU_time]
labels =['LSTM','GRU']
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

bplot1 = ax1.boxplot(niter,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax1.set_title('(a) Number of Iterations')

bplot2 = ax2.boxplot(time,
                     vert=True,  # vertical box alignment
                     patch_artist=True,  # fill with color
                     labels=labels)  # will be used to label x-ticks
ax2.set_title('(b) Total Computational Time')

# fill with colors
colors = ['pink', 'lightblue', 'lightgreen']
for bplot in (bplot1, bplot2):
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)

# adding horizontal grid lines
for ax in [ax1, ax2]:
    ax.yaxis.grid(True)
    ax.set_xlabel('Architectures')
#     ax.set_xlabel('Three architectures')
#     ax.set_ylabel('Observed values')

ax1.set_ylabel('epoch')
ax2.set_ylabel('seconds')
plt.show()
```
